```python
import matplotlib as mpl

def imsave_colormap(image, imname, colormap):
    """
    image -- numpy array from [0,1]
    imname -- string (filename to save result)
    colormap -- string (name of the colormap to save)
    """
    cm = mpl.cm.get_cmap(colormap)
    im = np.zeros(image.shape)
    im = cm(image)
    im = np.uint8(im * 255)
    im = Image.fromarray(im)
    im.save(imname)
```

### Calculate metrics

```python
model_name = 'clip50_l1-10.0_com-r-10.0_com-s-10.0_cycle-1.0'
#model_name = 'test'
```

```python
save_path = '/media/hpc-4_Raid/dsungatullina/submission/results/04.04.19/' + model_name + '/'

if not os.path.exists(save_path):
    os.mkdir(save_path)
```

```python
i = 0 
rmse_log_sum = 0.0
rmse_sum = 0.0
sqr_relative_sum = 0.0
abs_relative_sum = 0.0
accuracy_sum = 0.0

# baseline
rmse_log_sum_ss = 0.0
rmse_sum_ss = 0.0
sqr_relative_sum_ss = 0.0
abs_relative_sum_ss = 0.0
accuracy_sum_ss = 0.0

cols, rows = 2, 2

th = 1.25**3

for filename in r_filenames_img:
    
    img = Image.open(filename).convert('RGB')
    image = img_transform(img).unsqueeze(0) # torch tensor 1x3xhxw
    
    depth = cv2.imread(r_filenames_dpt[i], cv2.IMREAD_UNCHANGED)
    depth[depth > 50.0] = 50.0 # numpy array
    # depth[depth > 80.] = 80. # numpy array
    # depth = depth / 255. # [0,1]
    
    fake_depth_00 = netG_C_00(image.cuda())
    fake_depth_00 = fake_depth_00[4]
    fake_depth_00 = (fake_depth_00 + 1.0) / 2.0
    fake_depth_00 = fake_depth_00 * 255.
    fake_depth_00[fake_depth_00 > 50.] = 50.
    
    fake_depth_01 = netG_C_01(image.cuda())
    fake_depth_01 = fake_depth_01[4]
    fake_depth_01 = (fake_depth_01 + 1.0) / 2.0
    fake_depth_01 = fake_depth_01 * 255.
    fake_depth_01[fake_depth_01 > 50.] = 50.
    
    ar_fake_depth_00 = fake_depth_00.squeeze(0).squeeze(0).cpu().detach().numpy() # numpy array
    ar_fake_depth_01 = fake_depth_01.squeeze(0).squeeze(0).cpu().detach().numpy() # numpy array
    
    # unsupervised
    a_real = depth
    a_fake = ar_fake_depth_01
    mask1 = a_fake != 0
    mask2 = a_real != 0
    mask = mask1 * mask2
    volume1 = float(np.sum(mask1))
    volume2 = float(np.sum(mask2))
    volume = float(np.sum(mask))
    
    rmse_log = np.sqrt(np.sum((np.log(a_fake[mask]) - np.log(a_real[mask]))**2) / volume)
    rmse = np.sqrt(np.sum((a_fake - a_real)**2) / float(a_fake.shape[0]*a_fake.shape[1]))
    sqr_relative = np.sum((a_fake[mask2] - a_real[mask2])**2 / a_real[mask2]) / volume2
    abs_relative = np.sum(np.abs(a_fake[mask2] - a_real[mask2]) / a_real[mask2]) / volume2
    accuracy = np.sum(np.maximum(a_fake[mask]/a_real[mask], a_real[mask]/a_fake[mask]) < th) / volume
    
    rmse_log_sum = rmse_log_sum + rmse_log
    rmse_sum = rmse_sum + rmse
    sqr_relative_sum = sqr_relative_sum + sqr_relative
    abs_relative_sum = abs_relative_sum + abs_relative
    accuracy_sum = accuracy_sum + accuracy
    
    if i < 3:
        fig=plt.figure(figsize=(20, 10))
        print('Figure {}'.format(i))
        print('                           image                                                   groundtruth depth')
    
        fig.add_subplot(rows, cols, 1)
        plt.imshow(np.array(img))
        plt.axis('off')
        
        fig.add_subplot(rows, cols, 2)
        plt.imshow(depth)
        plt.axis('off')
        
        fig.add_subplot(rows, cols, 3)
        plt.imshow(ar_fake_depth_00)
        plt.axis('off')
        
        fig.add_subplot(rows, cols, 4)
        plt.imshow(ar_fake_depth_01)
        plt.axis('off')
        
        plt.savefig(save_path + '{}.png'.format(i))
        
        plt.show()
        print('                  supervised on synthetic                                             unsupervised')
        print()
        print('rmse_log: ', rmse_log)
        print('rmse: ', rmse)
        print('sqr_relative: ', sqr_relative)
        print('abs_relative: ', abs_relative)
        print('accuracy: ', accuracy)
        print()
    
    # supervised
    a_real = depth
    a_fake = ar_fake_depth_00
    mask1 = a_fake != 0
    mask2 = a_real != 0
    mask = mask1 * mask2
    volume1 = float(np.sum(mask1))
    volume2 = float(np.sum(mask2))
    volume = float(np.sum(mask))
    
    rmse_log = np.sqrt(np.sum((np.log(a_fake[mask]) - np.log(a_real[mask]))**2) / volume)
    rmse = np.sqrt(np.sum((a_fake - a_real)**2) / float(a_fake.shape[0]*a_fake.shape[1]))
    sqr_relative = np.sum((a_fake[mask2] - a_real[mask2])**2 / a_real[mask2]) / volume2
    abs_relative = np.sum(np.abs(a_fake[mask2] - a_real[mask2]) / a_real[mask2]) / volume2
    accuracy = np.sum(np.maximum(a_fake[mask]/a_real[mask], a_real[mask]/a_fake[mask]) < th) / volume
    
    rmse_log_sum_ss = rmse_log_sum_ss + rmse_log
    rmse_sum_ss = rmse_sum_ss + rmse
    sqr_relative_sum_ss = sqr_relative_sum_ss + sqr_relative
    abs_relative_sum_ss = abs_relative_sum_ss + abs_relative
    accuracy_sum_ss = accuracy_sum_ss + accuracy
    
    i = i + 1

print()
print('Average: (supervised vs unsupervised)')
print('rmse_log : ', rmse_log_sum_ss/float(i), rmse_log_sum/float(i))
print('rmse: ', rmse_sum_ss/float(i), rmse_sum/float(i))
print('sqr_relative: ', sqr_relative_sum_ss/float(i), sqr_relative_sum/float(i))
print('abs_relative: ', abs_relative_sum_ss/float(i), abs_relative_sum/float(i))
print('accuracy: ', accuracy_sum_ss/float(i), accuracy_sum/float(i))
```
